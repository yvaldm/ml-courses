```python
animals_dataset = pd.read_csv('https://stepik.org/media/attachments/course/4852/dogs_n_cats.csv')
```

```python
animals_dataset.head(5)
```

```python
animals_dataset = pd.get_dummies(animals_dataset)
 
```

```python
animals_dataset.head(5)
```

```python
animals_dataset = animals_dataset.drop(['Вид_собачка'], axis=1)
```

```python
animals_dataset.head(5)
```

```python
animals_dataset = animals_dataset.rename(index=str, columns={"Длина": "length", "Высота": "height", "Шерстист":"furry", "Гавкает":"bark", "Лазает по деревьям":"climb", "Вид_котик":"cat"})
```

```python
animals_dataset.head(5)
```

```python
X = animals_dataset.drop(['cat'], axis=1)
```

```python
y = animals_dataset['cat']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
scores_data = pd.DataFrame()
```

```python
max_depth_values = range(1, 50)


for max_depth in max_depth_values:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth, random_state=0)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    temp_score_data = pd.DataFrame({'max_depth':[max_depth], 'train_score':[train_score], 'test_score':[test_score]})
    scores_data = scores_data.append(temp_score_data)
    
```

```python
scores_data.head(25)
```

```python
scores_data_long = pd.melt(scores_data, id_vars=['max_depth'], value_vars=['train_score', 'test_score'],
                          var_name='set_type', value_name='score')
```

```python
scores_data_long
```

```python
sns.lineplot(x='max_depth', y='score', hue='set_type', data=scores_data_long)
```

```python
for max_depth in max_depth_values:
    clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score = clf.score(X_test, y_test)
    mean_cross_val_score = cross_val_score(clf, X_train, y_train, cv=5).mean()
    temp_score_data = pd.DataFrame({'max_depth':[max_depth], 'train_score':[train_score], 'test_score':[test_score],'mean_cross_val_score':[mean_cross_val_score]})
    scores_data = scores_data.append(temp_score_data)
```

```python
scores_data_long = pd.melt(scores_data, id_vars=['max_depth'], value_vars=['mean_cross_val_score','train_score', 'test_score'],
                          var_name='set_type', value_name='score')
```

```python
sns.lineplot(x='max_depth', y='score', hue='set_type', data=scores_data_long)
```
